## a. Identify and count missing values in a dataframe.

```python
var1 = dt.isnull().sum()
print(f'''\n--------------------------------------------------------------------
\nA.\n\nTotal null values in Given DataFrame : {sum(var1)}\n
{var1}
''')
```

## B. Dropping cols with more than 5 null

```python
var1 = [i for i in dt if var1[i] > 5]
print(f'''--------------------------------------------------------------------
\nB.\n\nColumns with more than 5 null values : {var1}\n
DataFrame after dropping columns : \n
{dt.drop(columns=var1).head()}
''')
```

## C. Dropping row with max row_sum value

```python
var1 = dt.sum(axis=1).idxmax()
print(f'''--------------------------------------------------------------------
\nC.\n\nRow with max sum value : {var1}\n
DataFrame after dropping row {var1} : \n
{dt.drop(index=var1)[(var1 - 2) : (var1 + 2)]}
''')
```

## D. Sorting DataFrame according to 1st col

```python
dt = dt.sort_values(by=dt.columns[0])
print(f'''--------------------------------------------------------------------
\nD.\n\nSorted dataFarme on the basis of first column : \n
{dt.head()}\n\n{dt.shape}
''')
```

## E. Removing duplicates from the 1st col

```python
dt.drop_duplicates(dt.columns[0], inplace=True)
print(f'''--------------------------------------------------------------------
\nE.\n\nDataFarme after removing duplicates from the first column :
\n {dt.head()}
\n{dt.shape}''')
```

## F. Calculating correlation & covariance

```python
var1 = dt.columns
print(f'''--------------------------------------------------------------------
\nF.\n\nCorrelation between first & second column : {dt[var1[0]].corr(dt[var1[1]])}
\nCovariance between second & third column : {dt[var1[1]].cov(dt[var1[2]])}
''')
```

## G. Detect & remove the row having outliers

```python
q = dt.quantile(q=[0.25, 0.75])
q.loc['IQR'] = q.iloc[1] - q.iloc[0]
q.loc['low'] = q.iloc[0] - 1.5 * q.iloc[2]
q.loc['high'] = q.iloc[1] + 1.5 * q.iloc[2]
dt = dt[~((dt < (q.loc['low'])) | (dt > (q.loc['high']))).any(axis=1)]
print(f'''--------------------------------------------------------------------
\nG.\n\nInter-Quartile Parameters for the outliers : \n\n{q}
\nDataFrame after removing outliers : \n\n{dt.head()}\n\n{dt.shape}
''')
```

## H. Discretizing second column & creating 5 bins

```python
dt['binned'] = pd.qcut(dt[dt.columns[1]], q=5)
var1 = dt.binned.unique()
print(f'''-------------------------------------------------------------------------
\nH.\n
Bins created for second column : \n
{var1.categories.values}
\nDataFrame after discretizing & creating 5 bins for second column : \n
{dt.head()}\n\n{dt.shape}''')
```
